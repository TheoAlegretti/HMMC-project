# Concerning the parametrization of the Gamma

```python
from particles import distributions as dists
import numpy as np
```

```python
kappa_s = 0.31

def mean_std(data):
    print(np.mean(data))
    print(np.std(data))
```

The values we want for the $\Gamma(\kappa_s, \frac{1}{\kappa_s})$

```python
print("mean:",1)
print("std:", 1/np.sqrt(kappa_s))
```

**First parametrization - reversing the second term**

```python
data = dists.Gamma(kappa_s, kappa_s).rvs(size = 10000)
mean_std(data)
```

Match !

**Second parametrization: not reversing the second term**

```python
data = dists.Gamma(kappa_s, 1/kappa_s).rvs(size = 10000)
mean_std(data)
```

Does not seem right
